# **Bài tập**

## **1. Mô tả**

- Cho một phương trình bậc 2 có dạng $y = ax^2 + bx + c = 0 , (a \neq 0)$. Hãy hoàn thành function giải phương trình bậc 2 trên, biết: $\Delta = b^2 - 4ac$

## **2. Ý tưởng: Tính nghiệm phương trình bậc 2 - làm quen với câu lệnh If/Else**

- Để tính được nghiệm, chúng ta dựa vào công thức sau đây:
  - Nếu $\Delta > 0$ -> phương trình có 2 nghiệm phân biệt là:
    - $x_1 = \frac{-b - \sqrt{\Delta}}{2a}$
    - $x_2 = \frac{-b + \sqrt{\Delta}}{2a}$
  - Nếu $\Delta = 0$ -> phương trình có 1 nghiệm kép $x_1 = x_2 = -\frac{b}{2a}$
  - Nếu $\Delta < 0$ -> phương trình vô nghiệm


- **Mở rộng:**: Trường hợp $a = 0$, hàm số sẽ thành phương trình bậc 1 có dạng: $y = bx + c = 0$, trường hợp này các bạn có thể mở rộng để bài toán tổng quát hơn.

  - Case 1:Phương trình vô số nghiệm:
    - Điều kiện: $a = 0$ và $b = 0$
  - Case 2: Phương trình vô nghiệm:
    - Điều kiện: $a = 0$ và $b \ne 0$
  - Case 3: Phương trình có nghiệm duy nhất:
    - Điều kiện: $a \ne 0$ và $b \ne 0$

```python
import math
```

```python
def calculate_delta(a: int,b: int,c:int):
  """
  Tính delta (Δ) của phương trình bậc hai:
    ax² + bx + c =

  Công thức:
    Δ = b² - 4a

  Parameters:
    a (int): Hệ số của x²
    b (int): Hệ số của x
    c (int): Hằng số tự d

  Returns:
    int: Giá trị delta của phương trình
  """

  # Áp dụng công thức tính delta
  delta = b**2 - 4*a*c

  return delta
```

```python
def solve_linear_equation(a: int,b: int)->str:
  """
  Giải phương trình bậc nhất dạng:
      ax + b = 0

  Parameters:
      a (int): Hệ số của x
      b (int): Hằng số tự do

  Returns:
      str: Kết quả của phương trình
  """

  # Sử dụng match-case để kiểm tra từng trường hợp
  # của cặp giá trị (a, b)
  match (a, b):
    # Trường hợp a = 0 và b = 0
    # => 0x + 0 = 0
    # => Phương trình vô số nghiệm với mọi x
    case (0, 0):
      return "Phương trình vô số nghiệm"

    # Trường hợp a = 0 nhưng b khác 0
    # Dấu "_" nghĩa là "mọi giá trị"
    # => 0x + b = 0
    # => vô nghiệm
    case (0, _):
      return "Phương trình vô nghiệm"

    # Các trường hợp còn lại:
    # a khác 0
    # => phương trình có nghiệm duy nhất
    case (a, b):
      x = -b / a
      return f"Phương trình có nghiệm duy nhất x = {x}"
```

```python
def solve_quadratic_equation(a: int,b: int, c:int):
  detal = calculate_delta(a,b,c)

  if detal < 0:
    return "Phương trình vô nghiệm"
  elif detal == 0:
    x = -b / (2*a)
    return f"Phương trình có nghiệm kép x1 = x2 = {x}"
  else:
    x1 = (-b + math.sqrt(detal)) / (2*a)
    x2 = (-b - math.sqrt(detal)) / (2*a)
    return f"Phương trình có 2 nghiệm phân biệt x1 = {x1}\n x2 = {x2}"
```

```python
def solve_equation(a:int, b:int, c:int):
  if a == 0:
    return solve_linear_equation(b,c)
  else:
    return solve_quadratic_equation(a,b,c)
```

```python
# Chạy thử các test case yêu cầu của bài
print("Test case 1:", solve_equation(a=2, b=6, c=4))
print("Test case 2:", solve_equation(a=1, b=2, c=1))
print("Test case 3:", solve_equation(a=4, b=6, c=3))
```
