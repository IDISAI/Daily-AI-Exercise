- Câu 1: Trình bày 2 cách (list, và enumerate) để đánh số thứ tự danh sách mua sắm.

```python
shopping_list = ["Cà rốt", "Táo", "Sữa"]

print("Danh sách mua sắm bằng list:")
for index in range(len(shopping_list)):
    print(f"{index + 1}. {shopping_list[index]}")

print("Danh sách mua sắm bằng enumerate:")
for index, item in enumerate(shopping_list, start=1):
    print(f"{index}. {item}")
```

- Câu 2: Tìm vị trí của các món trong list 2D - Trình bày 2 cách (list, và enumerate)

```python
food_list = [
    ["Bơ", "Pizza", "Sữa"],
    ["Xúc xích", "Táo", "Kem"],
    ["Cà rốt", "Bánh dâu", "Cupcake"]
]
search_items = ["Cà rốt", "Táo", "Sữa"]

def search_food_list(food_list, search_items):
  for i in range(len(food_list)):
    for j in range(len(food_list[i])):
      if(food_list[i][j]) in search_items:
        print (f'{food_list[i][j]} được tìm thấy ở hàng {i+1} và cột {j+1}')


print(search_food_list(food_list, search_items))
```

```python
def search_food_list_enumerate(food_list, search_items):
  for i, row in enumerate(food_list, start = 1):
    for j, item in enumerate(row):
      if item in search_items:
        print(f'{item} được tìm thấy ở hàng {i+1} và cột {j+1}')

print(search_food_list_enumerate(food_list, search_items))
```
